```python
# 라이브러리
import konlpy
from konlpy.tag import Okt
tokenizer = Okt()
```

```python
# 토큰화 ( 형태소 단위 )
text = "오늘 비가 와서 기분이 우울해요"
tokenizer.morphs(text)
```

```python
# 토큰화
tokenizer.pos(text)
```

### CountVectorizer

```python
# CountVectorizer
from sklearn.feature_extraction.text import CountVectorizer
vect = CountVectorizer()
```

```python
# 단어 토큰화
words = tokenizer.morphs(text)
```

```python
# 데이터 학습
vect.fit(words)
```

```python
# 학습된 어휘
vect.get_feature_names_out()
```

```python
# 단어 사전
vect.vocabulary_
```

```python
# 단어 사전 크기
len(vect.vocabulary_)
```

```python
# 인코딩
df_t = vect.transform(words)
```

```python
# 인코딩된 데이터 matrix
df_t.toarray()
```

```python
# 어휘와 피처 매칭 프레임
pd.DataFrame(df_t.toarray(), columns=vect.get_feature_names_out())
```

```python
# 새로운 단어가 들어오면 어떻게 매칭되는지
test = "날씨가 좋으면 기분이 좋아져요"
```

```python
# 단어 토큰화
words = tokenizer.morphs(test)
```

```python
# 인코딩된 데이터 matrix
test_t = vect.transform(words)
```

```python
test_t.toarray()
```

```python
# 어휘와 피처 매칭 프레임
pd.DataFrame(test_t.toarray(), columns=vect.get_feature_names_out())
```

### TfidfVectorizer

```python
# tf-idf
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
# tf-idf 활용한 어휘 사전 구축
vect = TfidfVectorizer()
words = tokenizer.morphs(text)
vect.fit(words)
vect.vocabulary_
```

```python
# 인코딩된 데이터 matrix
vect.transform(words).toarray()
```
